```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#  
df = pd.read_csv("./results/patch_size_results.csv")

# macro avg validation  
filtered_df = df[(df['class'] == 'macro avg') & (df['data_type'] == 'validation')]

#   
categories = filtered_df['Category'].unique()

# 
plt.figure(figsize=(12, 8))

for i, category in enumerate(categories, 1):
    plt.subplot(2, 2, i)
    
    #    
    cat_df = filtered_df[filtered_df['Category'] == category]
    
    #   (patch_size f1-score  )
    sns.lineplot(x=cat_df['patch_size'], y=cat_df['f1-score'], marker='o', label=category)
    
    plt.title(category)
    plt.xlabel('Patch Size')
    plt.ylabel('F1-Score')
    plt.xticks(sorted(cat_df['patch_size'].unique()))  # X 
    plt.grid(True)

plt.tight_layout()
plt.show()



# macro avg validation  
filtered_df = df[(df['class'] == 'accuracy') & (df['data_type'] == 'validation')]

#   
categories = filtered_df['Category'].unique()

# 
plt.figure(figsize=(12, 8))

for i, category in enumerate(categories, 1):
    plt.subplot(2, 2, i)
    
    #    
    cat_df = filtered_df[filtered_df['Category'] == category]
    
    #   (patch_size f1-score  )
    sns.lineplot(x=cat_df['patch_size'], y=cat_df['f1-score'], marker='o', label=category)
    
    plt.title(category)
    plt.xlabel('Patch Size')
    plt.ylabel('Accuracy')
    plt.xticks(sorted(cat_df['patch_size'].unique()))  # X 
    plt.grid(True)

plt.tight_layout()
plt.show()
```
